## Concepto de tuplas

"También se pueden identificar la tuplas, como un conjunto de valores agrupados
secuencialmente."

*-Izquierdo, J. (2020). Codifica en Python. Volumen I*

"Las tuplas son estructuras similares a las listas, pero con la diferencia principal de que son inmutables."

*-Gómez, J & García, J. (2025). Programación en Python. Más allá del código*

"Cada uno de los elementos que conforman una tupla puede ser de cualquier
tipo (básico o no). La sintaxis para declarar una tupla consiste en especificar sus elementos separados por una coma (,)."

*-Rodó, D. (2023). El lenguaje Python*
