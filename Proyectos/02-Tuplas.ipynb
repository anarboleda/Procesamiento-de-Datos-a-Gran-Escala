## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.

1. **Definición de tupla base (genres_tuple)**

- **Instrucciones :** genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", "R&B", "progressive rock", "disco")

- **Algoritmo:**

    1) Definir una colección ordenada de elementos (géneros) en una tupla.
    2) Almacenar la tupla en una variable para reutilizarla en consultas.
    3) Usar la tupla como fuente de datos para indexación, slicing y búsquedas.

2. **Longitud de la tupla (len(genres_tuple))**

- **Instrucciones :** len(genres_tuple)

- **Algoritmo:**

    1) Recibir la tupla genres_tuple.
    2) Contar la cantidad total de elementos contenidos.
    3) Retornar el número de elementos.

3. **Acceso por índice (genres_tuple[3])**

- **Instrucciones :** genres_tuple[3]

- **Algoritmo:**

    1) Recibir la tupla y el índice 3.
    2) Acceder al elemento ubicado en esa posición (indexación inicia en 0).
    3) Retornar el valor recuperado.

4. **Uso de slicing para sub-tupla (genres_tuple[3:6])**

- **Instrucciones :** genres_tuple[3:6]

- **Algoritmo:**

    1) Definir inicio = 3 y fin = 6.
    2) Extraer los elementos desde inicio (incluido) hasta fin (excluido).
    3) Retornar la sub-tupla resultante.
    4) Regla clave: el límite superior del slicing es excluyente.

5. **Primeros dos elementos (genres_tuple[0:2] o genres_tuple[:2])**

- **Instrucciones :** genres_tuple[0:2] o genres_tuple[:2]

- **Algoritmo:**

    1) Definir el rango para tomar desde el inicio hasta antes del índice 2.
    2) Extraer posiciones 0 y 1.
    3) Retornar la sub-tupla con esos elementos.

6. **Búsqueda de índice de un valor (genres_tuple.index("disco"))**

- **Instrucciones :** genres_tuple.index("disco")

- **Algoritmo:**

    1) Recorrer la tupla comparando cada elemento con "disco".
    2) Identificar la primera coincidencia exacta.
    3) Retornar el índice donde aparece.

- **Nota :** si el valor no existe, index() genera error.

7. **Ordenamiento de una tupla numérica (sorted(C_tuple))**

- **Instrucciones :** C_tuple = (-5, 1, -3), sorted(C_tuple)

- **Algoritmo:**

    1) Definir la tupla C_tuple con valores numéricos.
    2) Aplicar sorted() para ordenar ascendentemente los elementos.
    3) Retornar una lista nueva con los valores ordenados.

- **Nota :** sorted() devuelve lista, no tupla (si necesitas tupla: tuple(sorted(C_tuple))).

## Conclusiones

> En esta sección se presentarán las conclusiones obtenidas tras el desarrollo
de las actividades del cuaderno.


Las tuplas en Python son colecciones ordenadas e inmutables, por lo que son adecuadas cuando se requiere mantener datos constantes y solo consultarlos. La indexación empieza en 0 y el slicing usa un fin excluyente, lo cual debe tenerse presente para no cometer errores de rango. El método index() permite ubicar rápidamente un elemento, pero depende de que el valor exista (en escenarios reales conviene validar su presencia). Finalmente, sorted() no altera la tupla original y devuelve una lista, lo que evidencia que algunas operaciones retornan nuevos tipos de datos según el objetivo (consulta vs transformación).
