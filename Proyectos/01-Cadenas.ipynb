## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.


1. **Asignación e impresión (A, B)**
   - **Instrucciones :** `A = "1"`, `B = "2"`, `print(A)`, `print(B)`
   - **Algoritmo:**  
     1) Definir el literal de cadena.  
     2) Almacenar el literal en una variable.  
     3) Mostrar el valor por consola.

2. **Concatenación de cadenas (C = A + B)**
   - **Instrucciones :** `C = A + B`, `print(C)`
   - **Algoritmo:**  
     1) Tomar dos cadenas de entrada `A` y `B`.  
     2) Unirlas en el orden dado (sin conversión numérica).  
     3) Retornar la cadena resultante.  
   - **Nota :** si se requiere suma numérica, se tiene que convertir primero a numerico: `int(A) + int(B)`.

3. **Uso de slicing (D[0:3])**
   - **Instrucciones :** `D = "ABCDEFG"`, `print(D[0:3])` o `D[:3]`
   - **Algoritmo :**  
     1) Definir inicio = 0 y fin = 3.  
     2) Extraer caracteres desde inicio hasta fin.  
     3) Retornar la subcadena.  
   - **Regla clave:** la indexación de cadenas empieza en 0.

4. **Uso de stride (E[::2])**
   - **Instrucciones :** `E = "clocrkr1e1c1t"`, `print(E[::2])`
   - **Algoritmo:**  
     1) Iniciar en la primera posición de la cadena.  
     2) Tomar un carácter y avanzar con paso 2 (posiciones 0, 2, 4, …).  
     3) Retornar la nueva cadena con los caracteres seleccionados.

5. **Impresión de barra invertida (\)**
   - **Instrucciones :** `print("\\")` o `print(r" \ ")`
   - **Algoritmo :**  
     1) Interpretar `\\` como un solo carácter barra invertida literal.  
     2) Retornar el carácter.  

6. **Transformación a mayúsculas (F.upper())**
   - **Instrucciones :** `F = "You are wrong"`, `print(F.upper())`
   - **Algoritmo:**  
     1) Recorrer cada carácter de la cadena.  
     2) Convertir letras a su equivalente en mayúscula.  
     3) Retornar una nueva cadena convertida.  

7. **Búsqueda de subcadena (G.find("snow"))**
   - **Instrucciones :** `G = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb Its fleece was white as snow"`
     y `print(G.find("snow"))`
   - **Algoritmo:**  
     1) Recorrer `G` buscando la primera coincidencia del patrón.  
     2) Si existe, retornar el índice de inicio; si no, retornar `-1`.

8. **Reemplazo de subcadena (H = G.replace("Mary", "Bob"))**
   - **Instrucciones :** `H = G.replace("Mary", "Bob")`, `print(H)`
   - **Algoritmo:**  
     1) Localizar ocurrencias del patrón `"Mary"`.  
     2) Sustituir cada ocurrencia por `"Bob"`.  
     3) Retornar una nueva cadena con el resultado.

## Conclusiones

> En esta sección se presentarán las conclusiones obtenidas tras el desarrollo
de las actividades del cuaderno.


- En Python, una **cadena** es una secuencia ordenada e inmutable de caracteres: cualquier “modificación” (como `upper()` o `replace()`) produce una nueva cadena.
- La **concatenación** con `+` no realiza operaciones aritméticas; solo une texto. Si se requieren cálculos, primero debe hacerse una conversión de tipo(por ejemplo, `int()`).
- La **indexación** inicia en 0, por lo que el primer carácter está en la posición `0`. Esto afecta directamente los rangos usados en **slicing**.
- El **slicing** es una herramienta directa para extraer subcadenas 
- el **stride** es usado para tomar patrones, lo cual permite filtrar caracteres.
- Para representar caracteres especiales como la **barra invertida** es necesario comprender las **secuencias de escape** y cómo “escapar” el carácter `\`.
- Los métodos como `upper()`, `find()` y `replace()` simplifican tareas frecuentes de procesamiento.
