## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.

1. **Creación de lista heterogénea (a_list)**

- **Instrucciones :** a_list = [1, "hello", [1, 2, 3], True]

- **Algoritmo:**

    1) Definir una lista con elementos de diferentes tipos.
    2) Incluir una lista anidada como elemento interno.
    3) Almacenar la estructura en a_list para consultas posteriores.

2. **Acceso por índice (a_list[1])**

- **Instrucciones :** a_list[1]

- **Algoritmo:**

    1) Recibir la lista a_list y el índice 1.
    2) Acceder al elemento en la posición indicada (indexación inicia en 0).
    3) Retornar el elemento recuperado.

3. **Uso de slicing en listas (a_list[1:4])**

- **Instrucciones :** a_list[1:4]

- **Algoritmo:**

    1) Definir inicio = 1 y fin = 4.
    2) Extraer elementos desde inicio (incluido) hasta fin (excluido).
    3) Retornar la sublista resultante.
    4) Regla clave: el límite superior del slicing es excluyente.

4. **Concatenación de listas (A + B)**

- **Instrucciones :** A = [1, "a"], B = [2, 1, "d"], A + B

- **Algoritmo:**

    1) Definir las listas de entrada A y B.
    2) Unirlas en orden usando el operador +.
    3) Retornar una nueva lista con todos los elementos combinados.

- **Nota :** la concatenación no modifica A ni B; genera otra lista.

Las listas en Python son colecciones ordenadas y mutables, lo que permite usarlas cuando se necesita agregar, eliminar o modificar elementos durante el procesamiento. Soportan múltiples tipos de datos y estructuras anidadas, lo que las hace útiles para modelar información compleja. La indexación inicia en 0 y el slicing mantiene el fin excluyente, igual que en cadenas y tuplas.
