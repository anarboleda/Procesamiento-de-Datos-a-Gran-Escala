## Concepto de diccionarios

"Los diccionarios o arrays asociativos, son estructuras de datos que asocian un valor a una clave (es única), pero un mismo valor puede ser asociado a diferntes claves"

*-Gortázar,F & Martínez, R & Fresno, V. (2016). Lengiajes de Programación y procesadores*

"Los diccionarios, datos de un tipo que, asocian objetos en pares."

*-Izquierdo, J. (2020). Codifica en python*

**Referencias:**

-Gortázar,F & Martínez, R & Fresno, V. (2016). *Lengiajes de Programación y procesadores*

## Ejemplo de sintaxis de diccionarios

```python
persona = {
    "nombre": "Ana",
    "edad": 21
}

persona["nombre"]
```

**Ideas clave**

-Almacenan datos en pares clave–valor.

-Permiten acceso rápido mediante la clave.

-No permiten claves duplicadas.

## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.

1. **Definición de diccionario base (soundtrack_dic)**

- **Instrucciones :** soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}

- **Algoritmo:**

    1) Definir una estructura de pares clave–valor donde la clave identifica el elemento.
    2) Asociar cada película (clave) con su año (valor).
    3) Guardar el diccionario para consultas posteriores.

2. **Obtener llaves del diccionario (soundtrack_dic.keys())**

- **Instrucciones :** soundtrack_dic.keys()

- **Algoritmo:**

    1) Acceder al diccionario soundtrack_dic.
    2) Extraer el conjunto de claves mediante keys().
    3) Retornar la vista de claves.

- **Nota :** keys() retorna una vista (no una lista), aunque se puede convertir con list(...).

3. **Obtener valores del diccionario (soundtrack_dic.values())**

- **Instrucciones :** soundtrack_dic.values()

- **Algoritmo:**

    1) Acceder al diccionario soundtrack_dic.
    2) Extraer los valores asociados a las claves con values().
    3) Retornar la vista de valores.

4. **Crear diccionario de ventas (album_sales_dict)**

- **Instrucciones :** album_sales_dict = {"The Bodyguard":50, "Back in Black":50, "Thriller":65}

- **Algoritmo:**

    1) Definir las claves como nombres de discos.
    2) Definir los valores como ventas (en millones).
    3) Construir el diccionario para permitir consulta directa por clave.

5. **Consultar un valor por clave (album_sales_dict["Thriller"])**

- **Instrucciones :** album_sales_dict["Thriller"]

- **Algoritmo:**

    1) Recibir la clave "Thriller".
    2) Buscar la clave en el diccionario.
    3) Retornar el valor asociado (ventas).

- **Nota :** si la clave no existe, se produce error; alternativa segura: album_sales_dict.get("Thriller").

6. **Obtener claves de album_sales_dict (album_sales_dict.keys())**

- **Instrucciones :** album_sales_dict.keys()

- **Algoritmo:**

    1) Acceder al diccionario album_sales_dict.
    2) Extraer las claves con keys().
    3) Retornar la vista de claves.

7. **Obtener valores de album_sales_dict (album_sales_dict.values())**

- **Instrucciones :** album_sales_dict.values()

- **Algoritmo:**

    1) Acceder al diccionario album_sales_dict.
    2) Extraer los valores con values().
    3) Retornar la vista de valores.

## Conclusiones

> En esta sección se presentarán las conclusiones obtenidas tras el desarrollo
de las actividades del cuaderno.


Los diccionarios almacenan información en pares clave–valor, permitiendo acceso rápido a un dato a partir de su clave. Las claves deben ser únicas (si se repite una clave, el valor anterior se sobrescribe). Los métodos keys() y values() facilitan recuperar claves o valores como “vistas” del diccionario, útiles para recorridos y validaciones. Para consultar datos, acceder por dicc[clave] es directo, y get(clave) ayuda a evitar errores cuando no se tiene certeza de que la clave exista.
