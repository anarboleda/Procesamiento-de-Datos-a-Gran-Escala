## Concepto de bucles

"Conjunto de una serie de fenómenos u operaciones que se repiten ordenadamente."

*-Real Academia Española. (2014). Diccionario de la lengua española (23ª ed.)*

"Los bucles son estructuras de control de flujo que permiten repetir un
bloque de código un determinado o indeterminado número de veces.
Esencialmente existen dos tipos de bucles: for y while."

*-Rodó, D. (2023). El lenguaje Python*

> Los bucles también pueden ser llamados sentencias iterativas o ciclos.

**Referencias:**

-Real Academia Española. (2014). *Diccionario de la lengua española* (23ª ed.)

-Rodó, D. (2023). *El lenguaje Python*

## Ejemplo de sintaxis de bucles

```python
# Bucle for: recorre una secuencia conocida
for numero in [1, 2, 3]:
    print("Número:", numero)


# Bucle while: se ejecuta mientras la condición sea verdadera
contador = 1

while contador <= 3:
    print("Contador:", contador)
    contador += 1
```

**Ideas clave**

-Permiten repetir instrucciones.

-Se ejecutan sobre secuencias.

-Reducen la repetición de código

-El bucle for se usa cuando se conoce la cantidad de repeticiones.

-El bucle while se usa cuando la repetición depende de una condición.
