## Concepto de funciones

"Una función es un subprograma el cual se le transmite una lista de argumentos (que puede ser vacía) y que devuelve un valor"

*-Badenas, J & llopis, J.(2001). Curso práctico de programación en C y C++*

"Todos los lenguajes de programación ofrecen mecanismos que permiten encapsular una serie de operaciones de forma parametrizable, y devolver unos resultados determinados."

*-Rodó, D. (2023). El lenguaje Python*
