## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.

1. **Definición de función de suma (ejemplo de sintaxis: sumar(a, b))**

- **Instrucciones :** def sumar(a, b):, return a + b

- **Algoritmo:**

    1) Definir una función con dos parámetros (a, b).
    2) Ejecutar la operación a + b dentro de la función.
    3) Retornar el resultado con return.

2. **Actividad 1: función de división (div(a, b))**

- **Instrucciones :** def div(a, b):, return (a/b)

- **Algoritmo:**

    1) Recibir dos entradas (a como numerador y b como denominador).
    2) Calcular la división a / b.
    3) Retornar el resultado.

- **Nota :** si b = 0, Python genera un error (división por cero), por lo que en un caso real conviene validarlo antes.

3. Definición de función genérica de suma/concatenación (con(a, b))

- **Instrucciones :** def con(a, b):, return (a + b)

- **Algoritmo:**

    1) Recibir dos entradas (a, b).
    2) Aplicar el operador + según el tipo de dato (suma o concatenación).
    3) Retornar el resultado.

4. **Usar con para enteros o cadenas**

- **Instrucciones :** con(10, 20) (y para cadenas aplica igual: con("Hola ", "Mundo"))

- **Algoritmo:**

    1) Enviar dos valores del mismo tipo compatible a la función.
    2) Ejecutar a + b:
    3) Si son números → realiza suma aritmética.
    4) Si son cadenas (str) → realiza concatenación.
    5) Retornar el resultado.

5. **Usar con para concatenar listas o tuplas**

- **Instrucciones :** con(['J', 'M'], [0, 1]) (y para tuplas aplica igual: con((1,2), (3,4)))

- **Algoritmo:**

    1) Enviar dos secuencias del mismo tipo (lista con lista / tupla con tupla).
    2) Aplicar a + b para unir los elementos en orden, generando una nueva secuencia.
    3) Retornar la secuencia resultante.

- **Nota :** mezclar tipos incompatibles (por ejemplo, int con str) genera error.

Las funciones permiten encapsular lógica reutilizable, recibir parámetros y retornar resultados, lo que mejora la organización del código y evita repetición. Por esto, es importante validar compatibilidad de tipos y considerar casos de error, como la división por cero en div(a, b) o la suma entre tipos no compatibles.
