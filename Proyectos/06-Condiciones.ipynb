## Ejemplo de sintaxis de condiciones

```python
edad = 18

if edad >= 18:
    print("Es mayor de edad")
```

**Ideas clave**

-Permiten tomar decisiones en el programa.

-Evalúan expresiones booleanas.

-Ejecutan bloques de código según la condición.

## Ejecución de actividades

> En esta sección se desarrollarán los ejercicios propuestos en clase
relacionados con el uso de condiciones en Python.


1. Escribe una declaración if que determine si un disco musical tiene una calificación mayor a 8. Compruébalo utilizando el calificación del disco <b>“Back in Black”</b> el cual tiene un valor de 8.5. Si la condición es verdadera, imprime “"This album is Amazing!"”

```python
# Respuesta =
rating = 8.5
if rating > 8:
    print ("This album is Amazing!")
```

2. Escribe una declaración if-else que determine lo siguiente. Si la calificación es mayor que 8 imprime “this album is amazing”. Si es menor o igual a 8 que imprima “this album is ok”.

```python
# Respuesta =
rating = 8.5
if rating > 8:
    print ("this album is amazing")
else:
    print ("this album is ok")
```

3. Escribe una declaración if que determine si un disco fue lanzado antes de 1980 o en los años 1991 o 1993. Si la condición es verdadera imprime el año de lanzamiento.

```python
# Respuesta =
album_year = 1979

if album_year < 1980 or album_year == 1991 or album_year == 1993:
    print ("this album came out already")
```

## Documentación de instrucciones y algoritmos

> Aquí se documentarán las instrucciones utilizadas y los algoritmos
implementados para la resolución de los ejercicios del taller.

1. **Condición simple (mayoría de edad)**

- **Instrucciones :** edad = 18, if edad >= 18:, print("Es mayor de edad")

- **Algoritmo:**

    1) Definir el valor de entrada (edad).
    2) Evaluar la condición edad >= 18.
    3) Si es verdadera, ejecutar la acción asociada (imprimir el mensaje).

2. **Condición simple (rating > 8)**

- **Instrucciones :** rating = 8.5, if rating > 8:, print("This album is Amazing!")

- **Algoritmo:**

    1) Definir la calificación (rating).
    2) Comparar rating > 8.
    3) Si es verdadera, imprimir el mensaje.

3. **Decisión con dos caminos (if-else con rating)**

- **Instrucciones :** rating = 8.5, if rating > 8:, print("this album is amazing"), else:, print("this album is ok")

- **Algoritmo:**

    1) Definir rating.
    2) Evaluar rating > 8.
    3) Si es verdadera, ejecutar el bloque del if; en caso contrario, ejecutar el bloque del else.

4. **Condición compuesta con or (año de álbum)**

- **Instrucciones :** album_year = 1979, if album_year < 1980 or album_year == 1991 or album_year == 1993:, print("this album came out already")

- **Algoritmo:**

    1) Definir album_year.
    2) Evaluar la condición compuesta (se cumple si cualquiera de las comparaciones es verdadera).
    3) Si la condición resulta verdadera, imprimir el mensaje.
